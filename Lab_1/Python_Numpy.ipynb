## 1. Basics of Python

### 1(a) Sum of current and previous number

```python

nums = input("Enter numbers separated by space: ").strip().split()
nums = [float(x) for x in nums]

if len(nums) < 2:
    print("Need at least two numbers to compute current+previous sums.")
else:
    for i in range(1, len(nums)):
        print(f"{nums[i]} + {nums[i-1]} = {nums[i] + nums[i-1]}")
```

### 1(b) Multiplication tables for 11–20

```python

for n in range(11, 21):
    multiples = [n * i for i in range(1, 11)]
    print(f"Table of {n}: {multiples}")
```

### 1(c) First 15 terms of Fibonacci sequence

```python

n_terms = 15
fib = []
a, b = 0, 1
for _ in range(n_terms):
    fib.append(a)
    a, b = b, a + b

print(f"First {n_terms} Fibonacci terms:\n{fib}")
```

### 1(d) Check for leap year

```python

year = int(input("Enter year: ").strip())
is_leap = (year % 4 == 0 and (year % 100 != 0 or year % 400 == 0))
if is_leap:
    print("Yes, the given year is a leap year")
else:
    print("No, the given year is not a leap year")
```

### 1(e) Check if string contains digits and count them

```python

s = input("Enter a string: ")
digit_count = 0
for ch in s:
    if ch.isdigit():
        digit_count += 1

if digit_count:
    print(f"String contains digits. Total digits: {digit_count}")
else:
    print("No digits contained")
```

## 2. Python Data Structures (Dictionaries)

```python

# (a) Create empty dictionary
customer_transactions = {}

# (b) Add records
customer_transactions['Ankit'] = [1200, 5200, 75]
customer_transactions['Amit']  = [4500, 13000, 25]
customer_transactions['Ravi']  = [200, 150, 10000]

# (c) Display
print("Transaction histories:")
for name, txns in customer_transactions.items():
    print(f"{name}: {txns[0]}, {txns[1]}, {txns[2]}")

# (d) Total per customer
totals = {name: sum(txns) for name, txns in customer_transactions.items()}
print("\nTotal transaction volume per customer:")
for name, total in totals.items():
    print(f"{name}: {total}")

# (e) Highest total
max_customer = max(totals.items(), key=lambda x: x[1])
print(f"\nCustomer with highest total transactions: {max_customer[0]} ({max_customer[1]})")

# (f) Suspicious transactions (>10000 or <100)
suspicious_customers = {name: txns for name, txns in customer_transactions.items()
                        if any(t > 10000 or t < 100 for t in txns)}
print("\nSuspicious customers (txn >10000 or <100):")
print(suspicious_customers)
```

## 3. Basics of NumPy

```python

import numpy as np

# (a)
A = np.arange(1, 13).reshape(4, 3)
print("A =\n", A)
print("shape:", A.shape)
print("size:", A.size)
print("dtype:", A.dtype)

# (b)
second_col = A[:, 1]
subarray = A[:2, -2:]
print("\nSecond column:", second_col)
print("Subarray (first 2 rows, last 2 columns):\n", subarray)

# (c)
B = A.reshape(3, 4)
C = np.full_like(B, 5)
print("\nB =\n", B)
print("C =\n", C)
print("B + C =\n", B + C)
print("B - C =\n", B - C)
print("B * C =\n", B * C)
print("B / C =\n", B / C)

# (d)
row_sum = B.sum(axis=1)
col_sum = B.sum(axis=0)
print("\nRow-wise sums:", row_sum)
print("Column-wise sums:", col_sum)

# (e)
adder = np.array([10, 20, 30, 40])
B_plus = B + adder
count_greater_6 = np.sum(B_plus > 6)
print("\nB after adding [10,20,30,40]:\n", B_plus)
print("Count of elements > 6:", int(count_greater_6))

B_modified = B.copy()
B_modified[B_modified < 6] = 0
print("\nB with elements < 6 replaced by 0:\n", B_modified)
```

## 4. Basics of Matplotlib

```python

import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
city1 = [22, 25, 28, 30, 35, 40]
city2 = [18, 20, 25, 27, 30, 32]

# (a, b)
plt.figure(figsize=(8,4))
plt.plot(months, city1, marker='o', label='City 1')
plt.plot(months, city2, marker='s', label='City 2')
plt.xlabel('Month')
plt.ylabel('Avg Temperature (°C)')
plt.title('Monthly Average Temperatures: City1 vs City2')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# (c) May–June bar chart
months_mj = ['May', 'Jun']
city1_mj = [city1[4], city1[5]]
city2_mj = [city2[4], city2[5]]

x = np.arange(len(months_mj))
width = 0.35

plt.figure(figsize=(6,4))
plt.bar(x - width/2, city1_mj, width, label='City 1')
plt.bar(x + width/2, city2_mj, width, label='City 2')
plt.xticks(x, months_mj)
plt.xlabel('Month')
plt.ylabel('Avg Temperature (°C)')
plt.title('May & June Temperatures Comparison')
plt.legend()
plt.tight_layout()
plt.show()
```
